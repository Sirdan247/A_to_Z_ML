```python
# Polynomial Regression

# import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the dataset
dataset = pd.read_csv('Position_Salaries.csv')
# dataset.head()

X = dataset.iloc[:,1:2] # To be in matrix form
y = dataset.iloc[:, 2]
```

```python
# Fitting a Linear Regression for reference/ comparison purpose
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Fitting a Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X,lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

```python
# Visualising the Polynomial Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue') # Don't use the static variable X_poly
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

```python
# Visualising the Polynomial Regression results in higher resolution
X_grid = np.arange(min(X.values), max(X.values), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid,lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

```python
# Predicting a new result with Linear Regression
lin_reg.predict([[6.5]]) # 2D array
```

```python
# Predicting a new result with polynomial Regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]])) # 2D array
```
